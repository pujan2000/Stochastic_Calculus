```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
delta_t = 1/500
max_t = 3

Bt = 0
t = delta_t
Bt_list = []

while t <= max_t:
    Bt = Bt + np.sqrt(delta_t) * norm.rvs()
    Bt_list.append(Bt)
    t += delta_t

plt.plot(Bt_list)
```

## P(min Bt < -0.5) for t = [0,3]

```python
counter = 0
for i in range(1000):
    Bt = 0
    t = delta_t
    Bt_list = []
    while t <= max_t:
        Bt = Bt + np.sqrt(delta_t) * norm.rvs()
        Bt_list.append(Bt)
        t += delta_t
    if (min(Bt_list) < -0.5) == True:
        counter += 1

counter/1000
```

```python
counter = 0
for i in range(1000):
    Bt = 0
    t = delta_t
    Bt_list = []
    while t <= max_t:
        Bt = Bt + np.sqrt(delta_t) * norm.rvs()
        Bt_list.append(Bt)
        t += delta_t
    if ((Bt_list[750] > 0) and (Bt_list[-1] < 0)) == True:
        counter += 1

counter/1000
```
